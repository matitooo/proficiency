```python
import pandas as pd
```

```python
results_cat = ['graph','linear','mixed','sequential']
df_results = {}
for cate in results_cat:
    df =pd.read_csv('best_params_'+cate+'.csv')
    df_results[cate] = df
```

```python
dfs = []

for name, df in df_results.items():
    temp = df.copy() 
    temp["name"] = name      
    dfs.append(temp)

merged = pd.concat(dfs, ignore_index=True)
```

```python
df_sorted = merged.sort_values('best_score',ascending=False)
```

```python
input_col = ['senses+full text','subjects attributes','subjects attributes','senses+subject attributes(for graph building)','senses+subject attributes(for graph building)','subjects attributes','senses+full text','subjects attributes','senses+full text+subject attributes(for graph building)']
```

```python
df_sorted['input'] = input_col
```

```python
df_sorted = df_sorted.drop(columns='name')
```

```python
df_sorted['params'] = df_sorted['params'].astype('str')
```

```python
df_sorted.to_csv('final_results.csv')
```
