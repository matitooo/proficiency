```python
import json
import numpy as np
import yaml
```

```python
path = "parsed_essays.jsonl"
```

```python
with open('config/senses_config.yaml', 'r') as f:
    data = yaml.safe_load(f)
    senses = data['senses']
    n_senses = len(senses)
```

```python
obj = []
sense_sequences = []
with open(path, "r", encoding="utf-8") as f:
    for line in f:
        data = json.loads(line)
        obj.append(data)
        relations = data['relations']
        connections  = []
        for sentence in relations:
            connections.append(sentence['Sense'][0])
        sense_sequences.append(connections)
```

```python
def emb_vec(i,length):
    vec = np.zeros(length)
    vec[i] = 1
    return vec
```

```python
senses_dict = dict(zip(senses,[emb_vec(i,n_senses) for i in range(len(senses))]))
```

```python
senses_oh = []
for text in sense_sequences:
    oh_text = []
    for sense in text:
        oh_text.append(senses_dict[sense])
    senses_oh.append(np.array(oh_text))
```
